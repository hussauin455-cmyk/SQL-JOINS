## Retrieve Customers with at Least One Order

### Subtask:
Retrieve all customers who have placed at least one order by writing and executing an SQL query using `INNER JOIN` between the `Customers` and `Orders` tables.

**Reasoning**:
The tables have been created and populated with dummy data. The next step, as per the main task, is to retrieve customers who have placed at least one order using an `INNER JOIN`. I will provide the SQL query for this.



````python
sql_query = """
SELECT DISTINCT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.email
FROM
    Customers c
INNER JOIN
    Orders o ON c.customer_id = o.customer_id;
"""

query_workbench_instructions = """
**Instructions for MySQL Workbench:**
1. Open MySQL Workbench and connect to your desired MySQL server instance.
2. Open a new SQL tab (File -> New Query Tab).
3. Copy and paste the SQL `SELECT` statement provided above into the query tab.
4. Execute the statement to retrieve the customers who have placed at least one order.
"""

print(sql_query)
print(query_workbench_instructions)
````

## Display All Customers and Orders (Matched or Not)

### Subtask:
Write and display an SQL query to retrieve all customers and all orders, including those that do not have a match in the other table. Since MySQL does not directly support `FULL JOIN`, this will be achieved by combining `LEFT JOIN` and `RIGHT JOIN` with `UNION`.

**Reasoning**:
To simulate a `FULL OUTER JOIN` in MySQL, which is necessary to retrieve all customers and all orders regardless of a match, I will use a combination of `LEFT JOIN` and `RIGHT JOIN` results and then combine them using `UNION`. The `LEFT JOIN` will ensure all customers are included, and the `RIGHT JOIN` will ensure all orders are included. `UNION` will remove duplicate rows where there is a match.


````python
sql_full_outer_join_query = """
SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.email,
    o.order_id,
    o.order_date,
    o.total_amount
FROM
    Customers c
LEFT JOIN
    Orders o ON c.customer_id = o.customer_id

UNION

SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.email,
    o.order_id,
    o.order_date,
    o.total_amount
FROM
    Customers c
RIGHT JOIN
    Orders o ON c.customer_id = o.customer_id
WHERE
    c.customer_id IS NULL; -- This condition ensures we only pick orders that were not matched by the LEFT JOIN
"""

full_outer_join_workbench_instructions = """
**Instructions for MySQL Workbench:**
1. Open MySQL Workbench and connect to your desired MySQL server instance.
2. Open a new SQL tab (File -> New Query Tab).
3. Copy and paste the SQL `SELECT` statement provided above into the query tab.
4. Execute the statement to retrieve all customers and all orders, whether matched or not. Pay attention to NULL values in customer or order columns, which indicate no match.
"""

print(sql_full_outer_join_query)
print(full_outer_join_workbench_instructions)
````

## Retrieve Customers Who Made Payments but Did Not Place Any Orders

### Subtask:
Write and display an SQL query to retrieve customers who have made payments but have not placed any orders. This will be achieved by joining `Customers` with `Payments` and then using a `LEFT JOIN` with `Orders`, filtering for `NULL` order IDs from the `Orders` table.

**Reasoning**:
To find customers who made payments but no orders, I will perform an `INNER JOIN` between `Customers` and `Payments` to get all customers who have paid. Then, I will `LEFT JOIN` this result with the `Orders` table. Finally, I will filter the results to include only those records where the `order_id` from the `Orders` table is `NULL`, indicating that the customer has made a payment but does not have any associated order.

````python
sql_customers_paid_no_orders_query = """
SELECT DISTINCT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.email
FROM
    Customers c
INNER JOIN
    Orders o_payments ON c.customer_id = o_payments.customer_id
INNER JOIN
    Payments p ON o_payments.order_id = p.order_id
LEFT JOIN
    Orders o_orders ON c.customer_id = o_orders.customer_id
WHERE
    o_orders.order_id IS NULL;
"""

customers_paid_no_orders_workbench_instructions = """
**Instructions for MySQL Workbench:**
1. Open MySQL Workbench and connect to your desired MySQL server instance.
2. Open a new SQL tab (File -> New Query Tab).
3. Copy and paste the SQL `SELECT` statement provided above into the query tab.
4. Execute the statement to retrieve the list of customers who have made payments but have not placed any orders.
"""

print(sql_customers_paid_no_orders_query)
print(customers_paid_no_orders_workbench_instructions)
````
